## I — Formulation CP-SAT

### Variables

Pour chaque chirurgie $i$ :

- $\text{start}_i \in [\text{release}_i, \text{horizon}]$ : entier (minutes)
- $\text{end}_i = \text{start}_i + \text{duration}_i$
- $\text{itv}_i = \text{IntervalVar}(\text{start}_i, \text{duration}_i, \text{end}_i)$
- $\text{room}_{i,r} \in \{0,1\}$ : booléens sur les salles $r$ compatibles
- $\text{surg}_{i,s} \in \{0,1\}$ : booléens sur les chirurgiens $s$ compatibles

### Contraintes dures

1. **Affectation unique** : $\sum_r \text{room}_{i,r} = 1$ et $\sum_s \text{surg}_{i,s} = 1$.
2. **NoOverlap par salle** sur les intervalles optionnels élargis du temps de nettoyage $\text{clean}_r$.
3. **Cumulative capacité 1 par chirurgien** sur les intervalles optionnels par chirurgien.
4. **Cumulative par équipement** : pour chaque type d'équipement $e$, somme des demandes simultanées $\leq$ capacité du pool.
5. **Deadline urgences** : $\text{start}_i \leq \text{deadline}_i$.
6. **Shift chirurgien** : $\text{start}_i \geq \text{shift}^{\text{start}}_s$ et $\text{end}_i \leq \text{shift}^{\text{end}}_s$ si $\text{surg}_{i,s} = 1$.

### Objectif (soft)

$$\min\; \alpha \cdot \text{makespan} + \beta \sum_i w_i (\text{start}_i - \text{release}_i) + \gamma \sum_i \text{pref\_pen}_i$$

avec $w_i = $ priorité (URGENT=5, ELECTIVE=1) et $\text{pref\_pen}_i = 1$ si le chirurgien préféré n'est pas affecté.

## II — Génération d'une instance

Le générateur (`src/instances.py`) suit les caractéristiques classiques d'un bloc hospitalier :

- 5 spécialités (cardio, ortho, neuro, general, uro), allocation du nombre par spécialité bornée par la capacité chirurgien.
- ~20 % d'urgences (avec deadline ≥ 2× durée), au plus une à deux par spécialité.
- Salles partiellement spécialisées (2 à 3 spécialités), temps de nettoyage 10–20 min.
- Chirurgiens mono- ou bispécialistes avec fenêtre horaire.
- Pool d'équipements mobiles (microscope, robot, échographe).

````python
INSTANCE_SEED = 42
instance = generate_instance(
    n_surgeries=14, n_rooms=3, n_surgeons=4, seed=INSTANCE_SEED
)
print(format_instance(instance))
````

## III — Résolution exacte par CP-SAT

````python
r_cp = solve_cp_sat(instance, time_limit_s=10.0)
print(f'Status   : {r_cp.status}')
print(f'Makespan : {r_cp.makespan} min  ({r_cp.makespan//60} h {r_cp.makespan%60:02d})')
print(f'Attente  : {r_cp.total_wait} min (pondérée)')
print(f'Objectif : {r_cp.objective:.0f}')
print(f'Temps    : {r_cp.solve_ms:.0f} ms')
print()
print(format_schedule(r_cp, instance))
````

## IV — Baseline gloutonne (règle de priorité)

Heuristique de dispatch combinant :

1. **Priorité** : urgents avant électifs.
2. **Deadline** (EDD) puis **durée** (SPT) pour les électifs.
3. Plusieurs règles de tri tentées (EDD-SPT, EDD-LPT, FCFS) + restarts aléatoires.
4. À chaque chirurgie, on choisit le couple (salle, chirurgien) qui démarre le plus tôt, pénalisé par un coût d'opportunité pour éviter de bloquer les ressources rares (ex. la seule salle cardio).

````python
r_pr = solve_priority_rule(instance)
print(f'Status   : {r_pr.status}')
print(f'Makespan : {r_pr.makespan} min')
print(f'Attente  : {r_pr.total_wait} min')
print(f'Objectif : {r_pr.objective:.0f}')
print(f'Temps    : {r_pr.solve_ms:.1f} ms')

plot_gantt(r_pr, instance)
plt.savefig('figures/03_gantt_priority.png', dpi=130, bbox_inches='tight')
plt.show()
````

## V — Recuit simulé

Le recuit simulé manipule la **permutation** des chirurgies passée au décodeur glouton. À chaque itération on perturbe par swap ou relocation, et on accepte les mouvements avec la probabilité de Metropolis $e^{-\Delta/T}$, $T$ refroidi géométriquement.

L'état initial est l'ordre produit par la règle de priorité (en cas d'échec, recherche aléatoire d'une permutation faisable).

````python
r_sa = solve_simulated_annealing(instance, n_iter=5000, seed=1)
print(f'Status   : {r_sa.status}')
print(f'Makespan : {r_sa.makespan} min')
print(f'Attente  : {r_sa.total_wait} min')
print(f'Objectif : {r_sa.objective:.0f}')
print(f'Temps    : {r_sa.solve_ms:.0f} ms')

plot_gantt(r_sa, instance)
plt.savefig('figures/04_gantt_sa.png', dpi=130, bbox_inches='tight')
plt.show()
````

## VII — Étude de scalabilité

On fait varier le nombre de chirurgies $n \in \{8, 12, 16, 20, 24\}$ en conservant le ratio salles/chirurgiens proportionnel, et l'on mesure l'objectif et le temps des trois solveurs.

CP-SAT reste compétitif tant que l'instance reste petite (~20 chirurgies) ; au-delà, sa supériorité s'estompe rapidement face à des heuristiques rapides.

````python
scale_rows = []
configs = [(8, 2, 3), (12, 3, 4), (16, 3, 5), (20, 4, 6), (24, 4, 7)]
for n_surg, n_rooms, n_surg_doctors in configs:
    for seed in range(3):  # 3 instances par taille
        inst = generate_instance(n_surgeries=n_surg, n_rooms=n_rooms,
                                 n_surgeons=n_surg_doctors, seed=10 + seed)
        pr = solve_priority_rule(inst, seed=seed)
        sa = solve_simulated_annealing(inst, n_iter=3000, seed=seed)
        cp = solve_cp_sat(inst, time_limit_s=8.0)
        scale_rows.append({
            'n': n_surg, 'seed': seed,
            'obj_pr': pr.objective if pr.status != 'INFEASIBLE' else None,
            'obj_sa': sa.objective if sa.status != 'INFEASIBLE' else None,
            'obj_cp': cp.objective if cp.status != 'INFEASIBLE' else None,
            'ms_pr': pr.solve_ms, 'ms_sa': sa.solve_ms, 'ms_cp': cp.solve_ms,
            'status_cp': cp.status,
        })
df_scale = pd.DataFrame(scale_rows)
df_scale.groupby('n').agg(
    obj_pr=('obj_pr', 'mean'), obj_sa=('obj_sa', 'mean'), obj_cp=('obj_cp', 'mean'),
    ms_pr=('ms_pr', 'mean'), ms_sa=('ms_sa', 'mean'), ms_cp=('ms_cp', 'mean'),
).round(1)
````

````python
agg = df_scale.groupby('n').mean(numeric_only=True)
fig, axes = plt.subplots(1, 2, figsize=(11, 3.5))
axes[0].plot(agg.index, agg['obj_pr'], 'o-', label='priority', color='#16A34A')
axes[0].plot(agg.index, agg['obj_sa'], 's-', label='SA', color='#D97706')
axes[0].plot(agg.index, agg['obj_cp'], '^-', label='CP-SAT', color='#2563EB')
axes[0].set_xlabel('Nombre de chirurgies')
axes[0].set_ylabel('Objectif moyen (minimisé)')
axes[0].set_title('Qualité de la solution')
axes[0].legend()

axes[1].plot(agg.index, agg['ms_pr'], 'o-', label='priority', color='#16A34A')
axes[1].plot(agg.index, agg['ms_sa'], 's-', label='SA', color='#D97706')
axes[1].plot(agg.index, agg['ms_cp'], '^-', label='CP-SAT', color='#2563EB')
axes[1].set_xlabel('Nombre de chirurgies')
axes[1].set_ylabel('Temps de résolution (ms)')
axes[1].set_yscale('log')
axes[1].set_title('Temps de résolution')
axes[1].legend()
fig.tight_layout()
plt.savefig('figures/06_scalabilite.png', dpi=130, bbox_inches='tight')
plt.show()
````

## VIII — Influence des poids de l'objectif

L'objectif a trois composantes : makespan, attente pondérée, pénalité de préférence chirurgien. En jouant sur le poids $\gamma$ associé aux préférences, on déplace le compromis entre planning compact et respect des affectations souhaitées.

````python
rows = []
for gamma in [0, 10, 30, 100, 300]:
    r = solve_cp_sat(instance, time_limit_s=8.0, gamma=gamma)
    pref_ok = sum(
        1 for a in r.assignments
        if instance.surgeries[a.sid].preferred_surgeon is not None
        and instance.surgeries[a.sid].preferred_surgeon == a.surgeon
    )
    pref_total = sum(
        1 for s in instance.surgeries if s.preferred_surgeon is not None
    )
    rows.append({
        'gamma': gamma,
        'makespan': r.makespan,
        'attente': r.total_wait,
        'preferences_ok': f'{pref_ok}/{pref_total}',
        'objectif': round(r.objective, 0),
    })
pd.DataFrame(rows).set_index('gamma')
````

## IX — Conclusion

### Bilan

| Critère | Priority rule | Simulated annealing | CP-SAT |
|---------|:-------------:|:-------------------:|:------:|
| Optimalité garantie | ✗ | ✗ | ✓ (à terme) |
| Rapidité | ✓ <1 ms | ✓ ~500 ms | (jusqu'à `time_limit_s`) |
| Robustesse à l'infaisabilité | Médiocre | Bonne | Excellente |
| Lisibilité du modèle | Élevée | Moyenne | Très élevée |

### Limites

1. **Anesthésistes non modélisés** — devraient être traités comme une seconde ressource cumulative.
2. **Multi-jour** — l'horizon est fixé à une journée ; étendre nécessite d'ajouter des contraintes de quotas hebdomadaires.
3. **Incertitudes** — durées réelles bruitées ; une approche robuste (chance-constraints) serait pertinente.
4. **Annulations / arrivées tardives** — pas de re-scheduling dynamique.

### Perspectives

- Intégrer un modèle stochastique de durée (cf. projet A4).
- Comparer à une formulation MILP (Gurobi) ou à un solveur dédié (HEXALY).
- Plug-in interactif (Streamlit) pour autoriser un planificateur à figer manuellement certaines affectations.
